# Exploratory Data Analysis on Iris Dataset

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For better plots
sns.set(style="whitegrid")
```

```python

# Load the Iris dataset
df = pd.read_csv("Iris.csv")
df.head()
```

```python

# Shape of dataset
df.shape
```

```python

# Dataset info
df.info()
```

```python

# Check for missing values
df.isnull().sum()
```

```python

# Statistical summary of dataset
df.describe()
```

```python

# Unique species in dataset
df['Species'].unique()
```

```python

# Distribution of species
sns.countplot(data=df, x="Species")
plt.title("Count of Species")
plt.show()
```

```python

# Histograms for all numeric columns
df.drop("Id", axis=1).hist(bins=20, figsize=(12,8))
plt.show()
```

```python

# Pairplot for visualization of all features
sns.pairplot(df.drop("Id", axis=1), hue="Species")
plt.show()
```

```python
Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.drop("Id", axis=1).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python

# Boxplots for each feature by species
features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']

plt.figure(figsize=(15,10))
for i, col in enumerate(features, 1):
    plt.subplot(2,2,i)
    sns.boxplot(x="Species", y=col, data=df)
    plt.title(f"{col} by Species")
plt.tight_layout()
plt.show()
```


### Insights & Observations
- The dataset contains **150 samples** equally distributed across 3 species.
- Petal length and petal width are strong indicators to differentiate between species.
- Sepal features overlap more, making them less discriminative.
- No missing values were found in the dataset.
